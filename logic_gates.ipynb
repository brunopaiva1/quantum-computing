```python
import qiskit
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi
from math import sqrt
import qiskit.quantum_info as qi
from qiskit.visualization import plot_histogram
```

## Criação de uma esfera de Bloch

```python
q = QuantumRegister(1, 'q')
circuit = QuantumCircuit(q)

α = 1/sqrt(2)
β = 1/sqrt(2)

ψ = [α, β]
circuit.initialize(ψ, q[0])

circuit.draw('mpl')
```

Representação de um qubit(bra-ket) definido no espaço de Hílbert

```python
ψ = qi.Statevector(circuit)
ψ
```

```python
probs = ψ.probabilities()
print(probs)
```

```python
counts = ψ.sample_counts(shots=100000)
print(counts)
```

```python
plot_histogram(counts)
```

```python
ψ.draw('bloch')
```

Aplicando Porta lógica de Pauli x

```python
qc1 = circuit
qc1.x(0)
qc1.draw('mpl')
```

```python
ψ = qi.Statevector(qc1)
ψ
```

```python
counts = ψ.sample_counts(shots=100000)
print(counts)
```

```python
plot_histogram(counts)
```

```python
ψ.draw('bloch')
```

Analise: esfera permanece a mesma devido ao estado superposição
